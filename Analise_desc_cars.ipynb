##Notemos que os preços são do tipo String. Assim, para obter a média precisamos do comando astype('float').

```python
precos=[]
for i in range(0,204):
  if cars['price'][i] is not '?':
    precos.append(cars['price'][i])
```

```python
df_precos=pd.DataFrame(precos)
```

```python
avg = df_precos.astype('float').mean()
```

##A média de preços dos carros é de:

```python
avg
```

##Vamos listar quais diferentes marcas de carro aparecem em nosso dataframe e quantas vezes cada uma delas aparece:

```python
dict_marcas={}
```

```python
for i in range(1,205):
  if cars['make'][i] not in dict_marcas:
    dict_marcas[cars['make'][i]]=1
  else:
    dict_marcas[cars['make'][i]]=dict_marcas[cars['make'][i]]+1
```

```python
dict_marcas
```

```python
len(dict_marcas)
```

```python
maxim = max(dict_marcas.values())
```

```python
lista_marcas=list(dict_marcas.keys())
len(lista_marcas)
```

```python
carro_max=[]
for j in range(0,22):
  if dict_marcas[lista_marcas[j]]==maxim:
    carro=lista_marcas[j]
    carro_max.append(carro)
    print(f'A marca mais frequente no dataframe é {carro}, tendo aparecido {maxim} vezes.' )
```
